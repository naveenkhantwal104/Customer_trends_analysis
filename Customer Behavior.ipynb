```python
import pandas as pd
df=pd.read_csv('/Users/thenaveenkhantwal/Downloads/customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby ('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
## Age Group
bins = [0,18,25,35,45,60,100]
labels = ['Teen','18-25','26-35','36-45','46-60','60+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

print(df['Age Group'].value_counts().to_string())
```

```python
# Gender distribution
print(df['Gender'].value_counts().to_string())
```

### kpi's

```python
#Total Revenue
print("Total Revenue:", df['Purchase Amount (USD)'].sum())
```

```python
#Average purchase amount
print("Average Purchase:", df['Purchase Amount (USD)'].mean())
```

```python
# Maximum amount
print("Highest Purchase:", df['Purchase Amount (USD)'].max())
```

```python
# Minimum amount
print("Lowest Purchase:", df['Purchase Amount (USD)'].min())
```

### Product analysis

```python
# Moat purchased items
print(df['Item Purchased'].value_counts().head(10).to_string())
```

```python
# Most popular categories
print(df['Category'].value_counts().to_string())
```

```python
#Revenue by category
print(df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False).to_string())
```

```python
# Average revenue by category
print(df.groupby('Category')['Purchase Amount (USD)'].mean().to_string())
```

## Location analysis

```python
# Top States by revenue
print(df.groupby('Location')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10).to_string())
```

```python
# Customer count by location
print(df['Location'].value_counts().head(10).to_string())
```

### Subscription analysis

```python
print(df['Subscription Status'].value_counts().to_string())
```

```python
# Revenue from Subscribers vs Non-Subscribers
print(df.groupby('Subscription Status')['Purchase Amount (USD)'].sum().to_string())
```

## Shipping analysis

```python
print(df['Shipping Type'].value_counts().to_string())
```

## Top 10 Customers

```python
print(df.groupby('Customer ID')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10).to_string())
```

### Which gender spends more

```python
print(df.groupby('Gender')['Purchase Amount (USD)'].mean().to_string())
```

### Which season generates highest revenue

```python
print(df.groupby('Season')['Purchase Amount (USD)'].sum().sort_values(ascending=False).to_string())
```

### Most preferred payment method

```python
print(df['Payment Method'].value_counts().to_string())
```
